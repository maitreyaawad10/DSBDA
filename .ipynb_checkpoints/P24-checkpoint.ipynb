# CREATING DATA SUBSETS FOR AMOUNT OF REGION AFFECTED

```python
not_affected_subset = df[df['area'] == 0]
```

```python
not_affected_subset
```

```python
partially_affected_subset = df[(df['area'] > 0) & (df['area'] <= 100)]
partially_affected_subset
```

```python
mostly_affected_subset = df[df['area'] > 100]
mostly_affected_subset
```

# MERGING TWO SUBSETS

```python
merged_subset = pd.concat([not_affected_subset, mostly_affected_subset])
merged_subset
```

# MELTING DATA TO LONG FORMAT

```python
df.head()
```

```python
melted_data = pd.melt(mostly_affected_subset, id_vars=['month', 'temp'], var_name='attribute' , value_name='area affected', value_vars='area')
melted_data
```

# CASTING DATA TO WIDE FORMAT

```python
casted_data = melted_data.pivot_table(index=['month', 'temp'], columns='attribute', values='area affected')
casted_data
```
